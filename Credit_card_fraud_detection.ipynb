```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
legit.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
df = pd.concat([legit_sample, fraud], axis=0)
```

```python
df['Class'].value_counts()
```

```python
df.info()
```

```python
X = df.drop(columns='Class', axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=54)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm = LogisticRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
prediction = lm.predict(X_test)
```

```python
trainp = lm.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print("Training Data Accuracy: ",accuracy_score(trainp, y_train)*100)
```

```python
print("Test Data Accuracy: ", accuracy_score(prediction, y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,prediction))
```
