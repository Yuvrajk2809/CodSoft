```python
from apyori import apriori
```

```python
rules = apriori(transactions, min_support = 0.005, min_confidence = 0.2, min_lift = 3, max_length  =4)
```

```python
rules1 = apriori(transactions, min_support = 0.005, min_confidence = 0.2, min_lift = 3, max_length  =4)
```

```python
df = pd.DataFrame(columns = ['Items add', 'items base', 'confidence','lift'])
```

```python
for i in range(len(rules)):
    ib = rules[i].ordered_statistics[0].items_base
    ia = rules[i].ordered_statistics[0].items_add
    c = rules[i].ordered_statistics[0].confidence
    l = rules[i].ordered_statistics[0].lift
    df.loc[i] = [ia, ib, c,l]
```
